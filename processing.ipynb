```python
#template to clean a file

#get names from .sql
column_names = ["id", 
              "p_estimate", 
              "p_estimate_moe", 
              "n_estimate", 
              "n_estimate_moe", 
              "base_population", 
              "sample_size", 
              "state",
              "gender",
              "disability_type",
              "age",
              "race",
              "education",
              "hispanic",
              "statistics"]

df = pd.read_csv("raw/acs_prevalence_only.csv", header=None, names=column_names)

#the last column is usually just the statistics indicator (all the same for a file), so we drop it
df = df.drop(columns=["statistics"])

#to new csv
df.to_csv("static/acs_prevalence.csv", index=False)
```
