```python
import pandas as pd
```

```python
#template to clean a file

#clean prevalence

#get names from .sql
column_names = ["id", 
              "p_estimate", 
              "p_estimate_moe", 
              "n_estimate", 
              "n_estimate_moe", 
              "base_population", 
              "sample_size", 
              "state",
              "gender",
              "disability_type",
              "age",
              "race",
              "education",
              "hispanic",
              "statistics"]

df = pd.read_csv("raw/acs_prevalence_only.csv", header=None, names=column_names)

#the last column is usually just the statistics indicator (all the same for a file), so we drop it
df = df.drop(columns=["statistics"])

#to new csv
df.to_csv("static/acs_prevalence.csv", index=False)
```

```python
#clean prevalence2
df = pd.read_csv("raw/acs_prevalence_v2_header.csv")

df = df[df['Inst_population']==0]
df.drop(columns=["disability_status", 'd_vet_rate', 'Inst_population', 'health_insurance', 'statistic'], inplace=True)
df.rename(columns={'cname': 'state'}, inplace=True)

df.to_csv("static/acs_prevalence2.csv", index=False)
```

```python
#clean employment

#get names from .sql
column_names = ["id", 
              "p_estimate", 
              "p_estimate_moe", 
              "n_estimate", 
              "n_estimate_moe", 
              "base_population", 
              "sample_size", 
              "state",
              "gender",
              "disability_status",
              "disability_type",
              "age",
              "race",
              "education",
              "hispanic",
              "statistics"]

df = pd.read_csv("raw/acs_employment_rate_only.csv", header=None, names=column_names)

#the last column is usually just the statistics indicator (all the same for a file), so we drop it
df = df.drop(columns=["statistics"])

#to new csv
df.to_csv("static/acs_employment_rate.csv", index=False)
```

```python
#clean employment 2
df = pd.read_csv("raw/acs_employment_rate_v2_header.csv")

df = df.drop(columns=["d_vet_rate", "Inst_population", "health_insurance", "statistic"])
#df["disability_status"].unique()
df.rename(columns={'cname': 'state'}, inplace=True)

df.to_csv("static/acs_employment_rate2.csv", index=False)
```

```python
#clean poverty

#get names from .sql
column_names = ["id", 
              "p_estimate", 
              "p_estimate_moe", 
              "n_estimate", 
              "n_estimate_moe", 
              "base_population", 
              "sample_size", 
              "state",
              "disability_status",
              "disability_type",
              "statistics"]

df = pd.read_csv("raw/acs_poverty.csv", header=None, names=column_names)

#the last column is usually just the statistics indicator (all the same for a file), so we drop it
df = df.drop(columns=["statistics"])

#to new csv
df.to_csv("static/acs_poverty.csv", index=False)
```
